Counting the frequency of words.

```python
def word_frequency(word, text):
  return text.count(word) / float(len(text))

word_frequency('a', text1)
```

Counting the Lexical Diversity.

```python
def lexical_diversity(text):
  return len(set(text)) / float(len(text))

lexical_diversity(text1)
```

# 1.4 Stopwords

```python
fdist.plot(50, cumulative=False)
```

Languages for which the stopwords are available.

```python
from nltk.corpus import stopwords

stopwords.fileids()
```

```python
stop = set(stopwords.words('english'))
print(stop)
```

```python
# Lowercase the text and remove the stopwords

text1_clean = [w.lower() for w in text1]
text1_clean = [w for w in text1_clean if w not in stop]
len(text1_clean)
```

Using **punctuation()** in Python string module to get the punctuations and remove it from the corpus.

```python
from string import punctuation

print(punctuation)
punct = punctuation + '—«»–„“‘’'
print(punct)
```

```python
text1_clean = [w.strip(punct) for w in text1_clean]
text1_clean = [w for w in text1_clean if w not in punct]
len(text1_clean)
```

```python
fdist1_clean = FreqDist(text1_clean)
fdist1_clean.most_common(5)
```

```python
fdist1_clean.plot(25, cumulative=False)
```

# 1.4 Hapaxes

Using hapaxes() to get the words used only once in the text.

```python
fdist1_clean.hapaxes()
```

Checking the words with 15 or more characters.

```python
vocab = set(text1_clean)

long_words = []

for word in vocab:
  if len(word) >= 15:
    long_words.append(word)

long_words
```

Words longer than 7 characters and occur more than 10 times in the text.

```python
[w for w in set(text1_clean) if len(w) >7 and fdist1_clean[w]>10]
```

Count for these words using **text1.count()** function.

```python
long_frequent_words = [w for w in set(text1_clean) if len(w) > 7 and fdist1_clean[w] > 10]
counts = [(word, text1_clean.count(word)) for word in long_frequent_words]
counts[:25]
```

Arranging them in descending order.

```python
sorted(counts, key=lambda x:x[1], reverse = True)[:25]
```
